```python
from datasets import load_dataset
import pandas as pd
import json
import os
```

```python
vsmec = load_dataset("viethq1906/UIT-VSMEC-Sentiment-Relabelled")
vsfc_aug = load_dataset("viethq1906/UIT-VSFC-emotion-augmentation")
```

```python
vsmec_train_df = vsmec['train'].to_pandas()
vsmec_valid_df = vsmec['validation'].to_pandas()
vsmec_test_df = vsmec['test'].to_pandas()

vsfc_aug_train_df = vsfc_aug['train'].to_pandas()
vsfc_aug_valid_df = vsfc_aug['validation'].to_pandas()
vsfc_aug_test_df = vsfc_aug['test'].to_pandas()
```

```python
merged_train_df = pd.concat([vsmec_train_df, vsfc_aug_train_df], ignore_index=True)
merged_valid_df = pd.concat([vsmec_valid_df, vsfc_aug_valid_df], ignore_index=True)
merged_test_df = pd.concat([vsmec_test_df, vsfc_aug_test_df], ignore_index=True)

# Shuffle the merged datasets
merged_train_df = merged_train_df.sample(frac=1).reset_index(drop=True)
merged_valid_df = merged_valid_df.sample(frac=1).reset_index(drop=True)
merged_test_df = merged_test_df.sample(frac=1).reset_index(drop=True)
```

```python
print(merged_train_df.shape)
print(merged_valid_df.shape)
print(merged_test_df.shape)
```

```python
# Save to JSON files
def save_to_json(df, filepath):
    records = []
    for _, row in df.iterrows():
        record = {
            "sentence": row['sentence'],
            "sentiment": int(row['sentiment'])  # Convert to int to ensure proper format
        }
        records.append(record)
    
    with open(filepath, 'w', encoding='utf-8') as f:
        json.dump(records, f, ensure_ascii=False, indent=4)
    
    print(f"Saved {len(records)} records to {filepath}")

res_dir = 'research/res'
save_to_json(merged_train_df, os.path.join(res_dir, 'train.json'))
save_to_json(merged_valid_df, os.path.join(res_dir, 'validation.json'))
save_to_json(merged_test_df, os.path.join(res_dir, 'test.json'))
```
